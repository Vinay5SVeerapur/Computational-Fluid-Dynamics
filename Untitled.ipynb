```python
import numpy as np
from numpy.linalg import inv
A=np.array([(300,-100,0,0,0),(-100,200,-100,0,0),(0,-100,200,-100,0),(0,0,-100,200,-100),(0,0,0,-100,300)])
b=np.array([(200*200)+500,500,500,500,(200*500)+500])
sol=np.dot(inv(A),b)
print(sol)
l=np.array([0,1,2,3,4])
import matplotlib.pyplot as plt
plt.plot(l,sol)
```
